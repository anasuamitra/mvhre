```python
import pandas as pd
import pickle as pkl
```

```python
train_df = pd.read_csv("train.txt", sep="\t", header=None, dtype = str)
valid_df = pd.read_csv("valid.txt", sep="\t", header=None, dtype = str)
test_df = pd.read_csv("test.txt", sep="\t", header=None, dtype = str)
```

```python
train_df
```

```python
entity_2id = {}
rel_2id = {}
```

```python
for index, row in train_df.iterrows():
    h_id, r_id, t_id = str(row[0]), str(row[1]), str(row[2])
    if h_id not in entity_2id:
        entity_2id[h_id] = len(entity_2id.keys())
    if t_id not in entity_2id:
        entity_2id[t_id] = len(entity_2id.keys())
    if r_id not in rel_2id:
        rel_2id[r_id] = len(rel_2id.keys())
        
for index, row in valid_df.iterrows():
    h_id, r_id, t_id = str(row[0]), str(row[1]), str(row[2])
    if h_id not in entity_2id:
        entity_2id[h_id] = len(entity_2id.keys())
    if t_id not in entity_2id:
        entity_2id[t_id] = len(entity_2id.keys())
    if r_id not in rel_2id:
        rel_2id[r_id] = len(rel_2id.keys())
        
for index, row in test_df.iterrows():
    h_id, r_id, t_id = str(row[0]), str(row[1]), str(row[2])
    if h_id not in entity_2id:
        entity_2id[h_id] = len(entity_2id.keys())
    if t_id not in entity_2id:
        entity_2id[t_id] = len(entity_2id.keys())
    if r_id not in rel_2id:
        rel_2id[r_id] = len(rel_2id.keys())
```

```python
entity_2id
```

```python
rel_2id
```

```python
entity_2idx = {v:k for k, v in entity_2id.items()}
rel_2idx = {v:k for k, v in rel_2id.items()}
```

```python
entity_2idx
```

```python
with open('entities.dict', 'w') as f:
    for k, v in entity_2idx.items():
        f.write(str(k) + "\t" + str(v) + "\n")
with open('relations.dict', 'w') as f:
    for k, v in rel_2idx.items():
        f.write(str(k) + "\t" + str(v) + "\n")
```
