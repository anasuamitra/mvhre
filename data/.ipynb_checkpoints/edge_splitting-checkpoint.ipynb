```python
    def get_node_type(node_id):
        global nodes
        for i in range(len(nodes['shift'])):
            node_idx = nodes['node_id2index'][node_id]
            if (node_idx >= nodes['shift'][i]) and (node_idx < nodes['shift'][i] + nodes['count'][i]):
                return i

    def list_to_sp_mat(li):
        global nodes
        data = [x[2] for x in li]
        i = [x[0] for x in li]
        j = [x[1] for x in li]
        return sp.coo_matrix((data, (i, j)), shape=(nodes['total'], nodes['total'])).tocsr()
```

```python
    def load_nodes(args, path_name="", file_name='node.dat'):
        """
        return nodes dict
        total: total number of nodes
        count: a dict of int, number of nodes for each type
        attr: a dict of np.array (or None), attribute matrices for each type of nodes
        shift: node_id shift for each type. You can get the id range of a type by
                    [ shift[node_type], shift[node_type]+count[node_type] )
        """
        nodes_df = pd.read_csv("node.dat", sep="\t", header=None)
        if len(nodes_df.columns) == 4:
            nodes_df.columns = ["id", "name", "type", "features"]
        elif len(nodes_df.columns) == 3:
            nodes_df.columns = ["id", "name", "type"]
        else:
            raise Exception("Too few information to parse!")
        ''' Sort the node types and reset dataframe indices '''
        df = nodes_df.groupby('type', sort=True).apply(lambda x: x.reset_index(drop=True)).drop('type',
                                                                                                axis=1).reset_index()

        nodes = {'total': 0, 'count': Counter(), 'shift': {}, 'node_id2index': {}, 'node_type2ids': defaultdict(set),
                 'attr': defaultdict(dict)}


        for index, row in df.iterrows():
            node_id, node_type = 0, 0
            if len(nodes_df.columns) == 4:
                node_id, node_name, node_type, node_attr = int(row["id"]), str(row["name"]), int(row["type"]), row[
                    "features"]
                if node_attr is not np.nan:
                    node_attr = list(map(float, node_attr.split(',')))
                    nodes['attr'][node_type][index] = np.array(node_attr)
                else:
                    nodes['attr'][node_type][index] = -1
            elif len(nodes_df.columns) == 3:
                node_id, node_name, node_type = int(row["id"]), str(row["name"]), int(row["type"])
                nodes['attr'][node_type][index] = -1
            nodes['node_id2index'][node_id] = index
            nodes['node_type2ids'][node_type].add(node_id)
            nodes['count'][node_type] += 1
            nodes['total'] += 1

        shift = 0
        for i in range(len(nodes['count'])):
            nodes['shift'][i] = shift
            shift += nodes['count'][i]

        return nodes
```

```python
    def load_links(args, path_name, file_name_link='link.dat', file_name_meta='meta.dat'):
        """
        return links dict
            total: total number of links
            count: a dict of int, number of links for each type
            meta: a dict of tuple, explaining the link type is from what type of node to what type of node
            data: a dict of sparse matrices, each link type with one matrix. Shapes are all (nodes['total', nodes['total'])
        """
        edges_df = pd.read_csv(file_name_link, sep="\t", header=None)
        if len(edges_df.columns) == 5:
            edges_df.columns = ["h_id", "t_id", "r_id", "link_weight", "link_features"]
        elif len(edges_df.columns) == 4:
            edges_df.columns = ["h_id", "t_id", "r_id", "link_weight"]
        elif len(edges_df.columns) == 3:
            edges_df.columns = ["h_id", "t_id", "r_id"]
        else:
            raise Exception("Too few information to parse!")
        meta_df = None
        if os.path.exists(file_name_meta):
            meta_df = pd.read_csv(os.path.join(file_name_meta), sep="\t", header=None)
            if len(meta_df.columns) == 2:
                meta_df.columns = ["r_id", "meta_features"]
            else:
                raise Exception("Too few information to parse!")

        ''' Check if the nodes in the edges are valid '''
        edges_df = edges_df.loc[(edges_df['h_id'].isin(nodes['node_id2index'].keys())) & (
            edges_df['t_id'].isin(nodes['node_id2index'].keys()))]
        df = edges_df.groupby('r_id', sort=True).apply(lambda x: x.reset_index(drop=True)).drop('r_id',
                                                                                                axis=1).reset_index()
        h_id, t_id, r_id, link_weight = 0, 0, 0, 1.0
        edges = {'total': 0, 'count': Counter(), 'shift': defaultdict(dict), 'meta': {}, 'attr': {}, 'edges_orig': list(),
                 'edges': defaultdict(list), 'mat': defaultdict(list), 'adj': None}
        for index, row in df.iterrows():
            if len(edges_df.columns) == 5:
                h_id, t_id, r_id, link_weight, link_features = int(row["h_id"]), int(row["t_id"]), int(
                    row["r_id"]), float(row["link_weight"]), row["link_features"]
                link_feat = list(map(float, link_features.split(',')))
                edges['attr'][r_id][index] = np.array(link_feat)
            elif len(edges_df.columns) == 4:
                h_id, t_id, r_id, link_weight = int(row["h_id"]), int(row["t_id"]), int(row["r_id"]), float(
                    row["link_weight"])
                edges['attr'] = None
            elif len(edges_df.columns) == 3:
                h_id, t_id, r_id = int(row["h_id"]), int(row["t_id"]), int(row["r_id"])
                edges['attr'] = None

            if r_id not in edges['meta']:
                h_type = get_node_type(h_id)
                t_type = get_node_type(t_id)
                edges['meta'][r_id] = dict()
                edges['meta'][r_id]["ht_type"] = (h_type, t_type)
                if meta_df is not None:
                    feat_list = list(meta_df.loc[meta_df["r_id"] == r_id]["meta_features"])
                    meta_feat = list(map(float, feat_list[0].split(',')))
                    edges['meta'][r_id]["meta_features"] = np.array(meta_feat)
                else:
                    edges['meta'][r_id] = None

            h_idx = nodes['node_id2index'][h_id]
            t_idx = nodes['node_id2index'][t_id]
            edges['mat'][r_id].append((h_idx, t_idx, link_weight))
            edges['edges'][r_id].append((h_idx, t_idx, link_weight))
            edges['edges_orig'].append((h_idx, t_idx, r_id, link_weight)) # (h_id, t_id, r_id, link_weight)
            edges['count'][r_id] += 1
            edges['total'] += 1

        shift = 0
        for i in range(len(edges['count'])):
            edges['shift'][i] = shift
            shift += edges['count'][i]
        new_data = {}
        for r_id in edges['mat']:
            new_data[r_id] = list_to_sp_mat(edges['mat'][r_id])
        edges['mat'] = new_data
        adjM = sum(edges['mat'].values())
        edges['adj'] = adjM

        return edges
```

```python
links = load_links(0, "", file_name_link='link.dat', file_name_meta='meta.dat')
links
```

```python
target_link_type = 0
adj = links['mat'][target_link_type]
g = dgl.DGLGraph(adj)
g = dgl.remove_self_loop(g)
g
```

```python
u, v = g.edges()
eids = np.arange(g.number_of_edges())
eids = np.random.permutation(eids)
test_size = int(len(eids) * 0.1)
train_size = g.number_of_edges() - test_size
test_pos_u, test_pos_v = u[eids[:test_size]], v[eids[:test_size]]
train_pos_u, train_pos_v = u[eids[test_size:]], v[eids[test_size:]]
```

```python
all_list = set(links['edges'][target_link_type])
test_list = set()
for u, v in list(zip(test_pos_u, test_pos_v)):
    test_list.add((u.item(), v.item(), 1.0))
    test_list.add((v.item(), u.item(), 1.0))
test_list
```

```python
train_list = all_list - test_list
train_list
```

```python
df = pd.DataFrame(set(test_list))
df[3] = df[2]
df[2] = target_link_type
df
```
